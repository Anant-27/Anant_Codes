● Use the scikit-learn library.

● Predict the house price using a single feature (e.g., number of rooms). 
● Train a simple linear regression model. 
● Plot the regression line using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

data = fetch_california_housing()
features = data.data
target = data.target

print(data.feature_names)
print(data.target_names if hasattr(data, 'target_names') else "Target")
```

```python
data
```

```python
# Use 'AveRooms' (3rd column) as the only input feature
room_input = features[:, [2]]
train_x, test_x, train_y, test_y = train_test_split(room_input, target, test_size=0.2, random_state=42)
regressor = LinearRegression()
regressor.fit(train_x, train_y)
output = regressor.predict(test_x)

# Display the learned slope and intercept
print(f"Coefficient: {regressor.coef_[0]:.4f}")
print(f"Intercept: {regressor.intercept_:.4f}")
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(room_input, target, color='blue', alpha=0.5)
plt.xlabel("Average Rooms")
plt.ylabel("House Price")
plt.title("Scatter Plot of Rooms vs Price")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(test_x, test_y, color='lightgreen', label='True Values', alpha=0.6)
plt.plot(test_x, output, color='darkorange', label='Fitted Line', linewidth=2)
plt.xlabel("Avg Rooms per Home")
plt.ylabel("Median Value ($100k)")
plt.title("Linear Fit: Rooms vs Price")
plt.legend()
plt.grid(True)
plt.show()
```
